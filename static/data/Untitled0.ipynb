```python
import pandas as pd
```

```python
df = pd.read_excel('./PC50107584.xlsx')
```

```python
df.iloc[:]
```

```python
id = 0
root = {
    'id': id,
    'name': 'root',
    'children': list()
}
current_root = root
parent = root
for row in df.iterrows():
    code = row[1][0]
    question = row[1][1]
    effective = row[1][2]
    tot = row[1][3]
    repond = row[1][4]
    try:
        if not pd.isnull(effective) and not pd.isnull(question):
            if not question.isupper() and not pd.isnull(code):
                node = {
                    'id': id,
                    'question': question,
                    'code': code,
                    'effecif': effective,
                    'total': tot,
                    'repondants': repond,
                }
                current_root['children'].append(node)
            elif question.isupper() and pd.isnull(code):
                node = {
                    'id': id,
                    'question': question,
    #                 'code': code,
                    'effecif': effective,
                    'total': tot,
                    'repondants': repond,
                    'children': list()
                }
                root['children'].append(node)
                current_root = node
    except Exception as e:
        print id, code, question, effective, tot, repond
        raise e
    id += 1
```
