```python
import pandas as pd
```

```python
# df = pd.read_excel('./PC50107584.xlsx')
df = pd.read_excel('./MC-469050499-2.xlsx')
```

```python
id = 0
root = {
    'id': id,
    'name': 'root',
    'children': list()
}
current_root = root
parent = root
clusters = dict()
nodes = list()
cluster = None
for row in df.iterrows():
    code = row[1][0]
    title = row[1][1]
    question = row[1][2]
    effective = row[1][3]
    tot = row[1][4]
    repond = row[1][5]
    try:
        if not pd.isnull(effective) and not pd.isnull(question):
            if not question.isupper() and not pd.isnull(code):
                node = {
                    'id': id,
                    'question': question,
                    'code': code,
                    'effecif': effective,
                    'total': tot,
                    'repondants': repond,
                    'cluster': cluster,
                    'title': title,
                }
                nodes.append(node)
            elif question.isupper() and pd.isnull(code) and isinstance(effective, int):
                node = {
                    'id': id,
                    'question': question,
    #                 'code': code,
                    'effecif': effective,
                    'total': tot,
                    'repondants': repond,
#                     'children': list()
                    'cluster': question,
                    'title': title,
                }
                cluster = question
                clusters[question] = node
    except Exception as e:
        print id, code, question, effective, tot, repond
        raise e
    id += 1
```

```python
import json
```

```python
with open('sample.json', 'w') as f:
    json.dump({'clusters':clusters, 'nodes':nodes}, f)
```

```python
df = pd.read_excel('./Lex Solution Micellaire Verbato.xlsx')
```

```python
df
```

```python
lex = dict()
for row in df.iterrows():
    id = row[1][0]
    if isinstance(id, int):
        codes = row[1][1]
        if isinstance(codes, unicode):
            codes = codes.split('-')
        elif isinstance(codes, (float, int)):
            codes = [codes]
        else:
            print codes, type(codes)
        answer = row[1][2]
        for code in codes:
            item = {'answer': answer, 'id': id}
            if code in lex:
                lex[code].append(item)
            else:
                lex[code] = [item]
```

```python
with open('lex.json', 'w') as f:
    json.dump(lex, f)
```
